```python
import gensim
from gensim.models import Word2Vec,KeyedVectors

model = KeyedVectors.load_word2vec_format('/kaggle/input/googlenewsvectorsnegative300/GoogleNews-vectors-negative300.bin',binary=True,limit=500000)
```

```python
model['man'].shape
```

```python
model['man']
```

```python
model.most_similar("man")
```

```python
model.most_similar("cricket")
```

```python
model.most_similar("facebook")
```

```python
model.similarity("man","woman")
```

```python
model.similarity("man","python")
```

```python
model.doesnt_match(['PHP','java','monkey'])
```

**We can also perform arithmetic
**

```python
vec = model['king'] - model['man'] + model['woman']
model.most_similar([vec])
```
